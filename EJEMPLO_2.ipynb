**En este paso se realiza la conversion de los datos tipo TEXTO o STRING a datos NUMERICOS**

Algunas características son datos nominales o categóricos.

Los modelos de red neuronal no manejan variables categóricas. Es necesario codificar las variables categóricas en variables numéricas.

Para realizar esta codificación podemos usar la clase LabelEncoder & OrdinalEncoder



**Primero**: codificamos los valores del tensor en números ordinales. Para ello use la clase OrdinalEnconder() de la librería Scikit-Learn.

```python
from sklearn.preprocessing import OrdinalEncoder
oc = OrdinalEncoder()
A = oc.fit_transform(A)
```

```python
print(A)
```

**Segundo**: codificamos los valores solo del tensor B en equivalencias binarias.

Para ello use la clase OneHotEncoder()de la librería Sckit-Learn.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
B = ohe.fit_transform(B).toarray()
```

```python
print(B)
```

A continuación normalice los valores del tensor . La normalización la puede realizar con la clase `StandardScaler()` de Scikit-Learn.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
A = sc.fit_transform(A)
```

```python
print(A.shape)
print(B.shape)
```

**Antes de crear el modelo de aprendizaje automatico, debemos dividir el conjunto de datos en dos: entrenamiento y pruebas.**

- Conjunto de datos de entrenamiento (70% de los datos): con el que vamos a entrenar la red neuronal.

- Conjunto de datos de prueba (30% de los datos): con el que vamos a probar qué tan bien o mal está clasificando la red neuronal.


```python
from sklearn.model_selection import train_test_split
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.3, random_state=42 )
```

```python
# Importar librerías para la red neuronal con Keras
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils.np_utils import to_categorical
```

Sin título.png

* Se define mediante la librería Keras, un modelo `Sequential` que se refiere al modelo de redes neuronales en donde las capas del modelo están organizadas de manera secuencial.

* La primera capa `Dense` representa la primera capa oculta de la red neuronal. Esta capa tiene 500 neuronas y recibe 24 neuronas de la capa de entrada mediante el parámetro `input_dim`.

* La capa de entrada son 24 neuronas que representan los valores de "riesgo_aseguramineto". Las neuronas de esta capa oculta tienen como función de activación `relu`.

* La segunda capa oculta tiene 25 neuronas con una función de activación la función `relu`.

* La capa de salida tiene 6 neuronas. Estas 6 neuronas representan los 6 posibles valores de la variable `B`. Esta capa también usará como función de activación 'softmax'.

* Se compila el modelo especificando las función de coste (parámetro       `loss =categorical_crossentropy`), algoritmo de optimización de la función de coste (parámetro `optimizer = rmsprop`) y métrica para evaluar el modelo (parámetro `metrics = categorical_accuracy`).

```python
# Modelos de capas secuenciales
model = Sequential()

# Una capa de entrada
model.add(Dense(500, input_dim=24, activation='relu'))

# Segunda Capa oculta
model.add(Dense(25, activation='relu'))

# Capa de salida
model.add(Dense(6, activation="softmax"))

# Compilación del modelo con sus parámetros
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
# Resumen del modelo
model.summary()
```

Entrenamos el modelo pasando el conjunto de datos de entrenamiento representado en los tensores `A_train` y `B_train`.

* Épocas o *epochs*: 200
* Tamaño del lote o *batch size*: 2

pasamos el parámetro `validation_data` para indicar en el proceso contra qué datos se debe validar la clasificación.



```python
num_epochs = 200
num_batch_size = 4
history = model.fit(A_train,
                    B_train,
                    epochs=num_epochs,
                    batch_size=num_batch_size,
                    validation_data=(A_test,B_test),
                    verbose=0)
```

Evaluamos el modelo con el método `evaluate` que nos proporciona el acceso a las métricas de evaluación. Obtenemos de manera rápida la medida de desempeño del modelo tanto para los datos de entrenamiento como para los datos de prueba.

```python
score = model.evaluate(A_train, B_train, verbose=0)
model_metrics = zip(model.metrics_names, score)
model_metrics = set(model_metrics)
print("Desempeño para el conjunto de entrenamiento:", model_metrics)

score = model.evaluate(A_test, B_test, verbose=0)
model_metrics = zip(model.metrics_names, score)
model_metrics = set(model_metrics)
print("Desempeño para el conjunto de pruebas:", model_metrics)
```

```python
print(history.history)
for key, value in history.history.items() :
    print (key)
```

**En este paso se realiza la conversion de los datos tipo TEXTO o STRING a datos NUMERICOS**

Algunas caracterÃ­sticas son datos nominales o categÃ³ricos.

Los modelos de red neuronal no manejan variables categÃ³ricas. Es necesario codificar las variables categÃ³ricas en variables numÃ©ricas.

Para realizar esta codificaciÃ³n podemos usar la clase LabelEncoder & OrdinalEncoder



**Primero**: codificamos los valores del tensor en nÃºmeros ordinales. Para ello use la clase OrdinalEnconder() de la librerÃ­a Scikit-Learn.

```python
from sklearn.preprocessing import OrdinalEncoder
oc = OrdinalEncoder()
A = oc.fit_transform(A)
```

```python
print(A)
```

**Segundo**: codificamos los valores solo del tensor B en equivalencias binarias.

Para ello use la clase OneHotEncoder()de la librerÃ­a Sckit-Learn.

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
B = ohe.fit_transform(B).toarray()
```

```python
print(B)
```

A continuaciÃ³n normalice los valores del tensor . La normalizaciÃ³n la puede realizar con la clase `StandardScaler()` de Scikit-Learn.

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
A = sc.fit_transform(A)
```

```python
print(A.shape)
print(B.shape)
```

**Antes de crear el modelo de aprendizaje automatico, debemos dividir el conjunto de datos en dos: entrenamiento y pruebas.**

- Conjunto de datos de entrenamiento (70% de los datos): con el que vamos a entrenar la red neuronal.

- Conjunto de datos de prueba (30% de los datos): con el que vamos a probar quÃ© tan bien o mal estÃ¡ clasificando la red neuronal.


```python
from sklearn.model_selection import train_test_split
A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.3, random_state=42 )
```

```python
# Importar librerÃ­as para la red neuronal con Keras
import keras
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.utils.np_utils import to_categorical
```

Sin tÃ­tulo.png

* Se define mediante la librerÃ­a Keras, un modelo `Sequential` que se refiere al modelo de redes neuronales en donde las capas del modelo estÃ¡n organizadas de manera secuencial.

* La primera capa `Dense` representa la primera capa oculta de la red neuronal. Esta capa tiene 500 neuronas y recibe 24 neuronas de la capa de entrada mediante el parÃ¡metro `input_dim`.

* La capa de entrada son 24 neuronas que representan los valores de "riesgo_aseguramineto". Las neuronas de esta capa oculta tienen como funciÃ³n de activaciÃ³n `relu`.

* La segunda capa oculta tiene 25 neuronas con una funciÃ³n de activaciÃ³n la funciÃ³n `relu`.

* La capa de salida tiene 6 neuronas. Estas 6 neuronas representan los 6 posibles valores de la variable `B`. Esta capa tambiÃ©n usarÃ¡ como funciÃ³n de activaciÃ³n 'softmax'.

* Se compila el modelo especificando las funciÃ³n de coste (parÃ¡metro       `loss =categorical_crossentropy`), algoritmo de optimizaciÃ³n de la funciÃ³n de coste (parÃ¡metro `optimizer = rmsprop`) y mÃ©trica para evaluar el modelo (parÃ¡metro `metrics = categorical_accuracy`).

```python
# Modelos de capas secuenciales
model = Sequential()

# Una capa de entrada
model.add(Dense(500, input_dim=24, activation='relu'))

# Segunda Capa oculta
model.add(Dense(25, activation='relu'))

# Capa de salida
model.add(Dense(6, activation="softmax"))

# CompilaciÃ³n del modelo con sus parÃ¡metros
model.compile(optimizer='rmsprop',
              loss='categorical_crossentropy',
              metrics=['categorical_accuracy'])
# Resumen del modelo
model.summary()
```

Entrenamos el modelo pasando el conjunto de datos de entrenamiento representado en los tensores `A_train` y `B_train`.

* Ã‰pocas o *epochs*: 200
* TamaÃ±o del lote o *batch size*: 2

pasamos el parÃ¡metro `validation_data` para indicar en el proceso contra quÃ© datos se debe validar la clasificaciÃ³n.



```python
num_epochs = 200
num_batch_size = 4
history = model.fit(A_train,
                    B_train,
                    epochs=num_epochs,
                    batch_size=num_batch_size,
                    validation_data=(A_test,B_test),
                    verbose=0)
```

Evaluamos el modelo con el mÃ©todo `evaluate` que nos proporciona el acceso a las mÃ©tricas de evaluaciÃ³n. Obtenemos de manera rÃ¡pida la medida de desempeÃ±o del modelo tanto para los datos de entrenamiento como para los datos de prueba.

```python
score = model.evaluate(A_train, B_train, verbose=0)
model_metrics = zip(model.metrics_names, score)
model_metrics = set(model_metrics)
print("DesempeÃ±o para el conjunto de entrenamiento:", model_metrics)

score = model.evaluate(A_test, B_test, verbose=0)
model_metrics = zip(model.metrics_names, score)
model_metrics = set(model_metrics)
print("DesempeÃ±o para el conjunto de pruebas:", model_metrics)
```

```python
print(history.history)
for key, value in history.history.items() :
    print (key)
```
